```python
import numpy as np
import matplotlib.pyplot as plt
from my_envs.biped import BipedEnv
```

```python
env = BipedEnv(render_mode='human')
obs, info = env.reset()
ts =[0]
obss = [obs]
for i in range(env.max_steps):
    action = env.action_space.sample()/3
    # action = np.array([1, 1])
    obs, reward, terminated, truncated, info = env.step(action)
    ts.append(env.data.time)
    obss.append(obs)
    if terminated or truncated:
        break
env.close()
ts = np.array(ts)
obss = np.array(obss)
```

```python
body_ws = obss[:,:3]

print(np.mean(body_ws,axis=0))
print(np.std(body_ws,axis=0))
plt.plot(ts,body_ws)
plt.show()
```
